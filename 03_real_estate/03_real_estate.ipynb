Напишем функцию, которая будет принимать строку с отсутствующим значением жилой площади и возвращать среднее значение по похожим объектам. По идее, общая площадь зависит от серии дома. Серия дома неизвестна, но ее косвенные признаки - этажность дома, количество комнат и город. Поэтому напишем функцию зависимости от этажности, количества комнат и города.

Сперва оценим уникальные значения комнат и населенных пунктов встречающихся в таблице. Заодно на глаз посмотрим наличие неявных дубликатов в населенных пунктах - во всяком случае, они не бросаются в глаза.

```python
data['rooms'].value_counts()
```

```python
data['locality_name'].sort_values().unique()
```

Составим сводную таблицу с данными о среднем количестве комнат для домов различной этажности в разных населенных пунктах. В таблице получилось много пропусков.

```python
pivot_rooms = data.pivot_table(index=['locality_name', 'floors_total'], columns='rooms', values='living_area', aggfunc='mean')
display(pivot_rooms)
```

```python
print(pivot_rooms.loc[('Волосово', 3.0)][1])
print(pivot_rooms.loc[('Волосово', 4.0)][1])
print(pivot_rooms.loc[('Волосово', 4.0)][2])
print(pivot_rooms.loc[('Волосово', 4.0)][3])
# проверяю, как оно работает
```

Напишем функцию `fillna_rooms` и применим ее к датасету.

```python
def fillna_rooms(city, floors, rooms):
    '''
    Находит в таблице pivot_rooms нужную медиану.
    '''
    try:
        return round(pivot_rooms.loc[(city, floors)][rooms], 2)
    except:
        return 0
```

```python
fillna_rooms('Волосово', 4, 1) # проверяю, как работает
```

```python
# запишем результат в новый столбец 
data['living_area_new'] = data.apply(lambda row: fillna_rooms(row['locality_name'], row['floors_total'], row['rooms']), axis=1)

# заменяем пропуски значениями из нового столбца в той же строке.
data['living_area'] = data['living_area'].fillna(data['living_area_new'])
```
